```python
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score, roc_auc_score)

# 성능평가를 위해 함수 작성 

def get_clf_eval(y_valid, pred):
    acc = accuracy_score(y_valid, pred)
    pre = precision_score(y_valid, pred)
    re = recall_score(y_valid, pred)
    f1 = f1_score(y_valid, pred)
    auc = roc_auc_score(y_valid, pred)
    
    return acc, pre, re, f1, auc
```

```python
from sklearn.metrics import confusion_matrix

def print_clf_eval(y_valid, pred):
    confusion = confusion_matrix(y_valid, pred)
    acc, pre, re, f1, auc = get_clf_eval(y_valid, pred)
    
    print("=> confusion matrix")
    print(confusion)
    print("===============")
    
    print("Accuracy: {0:.4f}, Precision: {1:.4f}".format(acc,pre))
    print("Recall: {0:.4f}, F1: {1:.4f}, AUC:{2:.4f}".format(re, f1, auc))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# LogisticRegression : precision, recall, f1 0으로 나옴
lr_clf = LogisticRegression(random_state=6, solver="liblinear")
lr_clf.fit(X_train, y_train)
lr_pred = lr_clf.predict(X_valid)

print_clf_eval(y_valid, lr_pred)
```

```python
# DecisionTreeClassifier : precision, recall, f1 0으로 나옴
dt_clf = DecisionTreeClassifier(random_state=6, max_depth=4)
dt_clf.fit(X_train, y_train)
dt_pred = dt_clf.predict(X_valid)

print_clf_eval(y_valid, dt_pred)
```

```python
def get_result(model, X_train, y_train, X_valid, y_valid):
    model.fit(X_train, y_train)
    pred = model.predict(X_valid)
    
    return get_clf_eval(y_valid, pred)
```

```python
def get_result_pd(models, model_names, X_train, y_train, X_valid, y_valid):
    col_names = ["accuracy", "precision", "recall", "f1", "roc_auc"]
    tmp = []
    
    for model in models:
        tmp.append(get_result(model, X_train, y_train, X_valid, y_valid))
        
    return pd.DataFrame(tmp, columns=col_names, index=model_names)
```

```python
# ROC곡선 그래프 함수 생성
from sklearn.metrics import roc_curve

def draw_roc_curve(models, model_names, X_valid, y_valid):
    plt.figure(figsize=(5, 5))
    
    for model in range(len(models)):
        pred = models[model].predict_proba(X_valid)[:, 1]
        fpr, tpr, thresholds = roc_curve(y_valid, pred)
        plt.plot(fpr, tpr, label=model_names[model])
        
    plt.plot([0, 1], [0, 1], "k--", label="random quess")
    plt.title("ROC")
    plt.legend()
    plt.grid()
    plt.show()
    
draw_roc_curve(models, model_names, X_valid, y_valid)
```

```python
# 이상치 제거하는 함수 생성
def get_outlier(df=None, column=None, weight=1.5):
    fraud= df[df["fraud_YN"]==1][column]
    quantile_25 = np.percentile(fraud.values, 25)
    quantile_25 = np.percentile(fraud.values, 75)
    
    iqr = quantile_75 = quantile_25
    iqr_weight = iqr * weight
    lowest_val = quantile_25 - iqr_weight
    highest_val = quantile_75 + iqr_weight
    
    outlier_index = fraud[(fraud < lowest_val) | (fraud > highest_val)].index
    
    return outlier_index
```

```python
# insure_cost에서는 최대치를 벗어나는 값이 3개 있었음
get_outlier(df=trainval_df, column="insure_cost", weight=1.5)
```

```python
# repair_cost에서는 최대치를 벗어나는 값이 2개 있었음
get_outlier(df=trainval_df, column="repair_cost", weight=1.5)
```
